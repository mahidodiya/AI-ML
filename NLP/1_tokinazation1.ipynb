```python
corpus="""Artificial Intelligence is transforming many industries.
Python is one of the most popular programming languages.
Tokenization splits text into smaller meaningful units.
Machine learning models learn patterns from data.
Students can practice NLP concepts using simple datasets."""
```

```python
#paragraph to sentance
from nltk.tokenize import sent_tokenize
sentance=sent_tokenize(corpus)
print(sentance)
```

```python
#paragraph to word
from nltk.tokenize import word_tokenize
token=word_tokenize(corpus)
print(token)
print(len(token))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
import string
from nltk.tokenize import TreebankWordTokenizer#class
tokenizer=TreebankWordTokenizer()#load class
token=tokenizer.tokenize(corpus)#method tokenize
print(token)
print(len(token))
word=[word for word in token if word not in string.punctuation]
print(word)
print(len(word))
```
