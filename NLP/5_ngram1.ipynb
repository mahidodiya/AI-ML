here, we create N-Gram programs , n-gram is a sequence of N consecutive words it can be unigram(N=1),bigram(N=2),trigram(N=3).ngram_range(minimum,maximum)

1.unigram+bigram

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = [
    "I love machine learning",
    "Machine learning is fun"
]

cv=CountVectorizer(ngram_range=(1,2),stop_words='english')
x=cv.fit_transform(text)

print(cv.vocabulary_)
print(x.toarray())
```

2.bigram+trigram

```python
text = [
    "I love machine learning",
    "Machine learning is fun"
]

cv=CountVectorizer(ngram_range=(2,3),stop_words='english')
x=cv.fit_transform(text)

print(cv.vocabulary_)
print(x.toarray())
```
