Examples of Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

In this below code we create bow of given text and print vocabulry, bow matrix, and remove stop words

```python
text = [
    "dog bites man",
    "man bites dog",
    "dog eats food",
    "the dog is running"
]

cv=CountVectorizer(stop_words='english')
x= cv.fit_transform(text)

print(cv.vocabulary_)
print(x.toarray())
```

In below code we convert text into bow and count word frequency

```python
text1 = [
    "apple apple banana",
    "banana orange apple"
]

cv1=CountVectorizer()
y=cv1.fit_transform(text1)

print(cv1.vocabulary_)
print(y.toarray())

word_count= y.toarray().sum(axis=0)
for word,index in cv1.vocabulary_.items():
    print(word , ":" , word_count[index])
```
