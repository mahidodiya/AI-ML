PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
words = ["playing","played","plays","walking","walked","walks","running","runner","runs","studying","studies","studied","happiness","happily","beautifully","beauty","organization","organizing","organized","connection","connected","connecting","connections","relational","relationship","relating","relativity","communication","communicating","communicated","communications","university","universities","universal","universally","arguing","argument","arguments","computational","computation","computing","computed","computer","computers","cats","catlike","fishing","fished","fisherman","fishes"]
```

```python
for word in words:
    print(word+"--->"+stemming.stem(word))
```

LANcaster stemming algorithm

```python
from nltk.stem import LancasterStemmer
```

```python
lancaster=LancasterStemmer()
```

```python
for word in words:
    print(word+"--->"+lancaster.stem(word))
```

Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowball=SnowballStemmer('english',ignore_stopwords=False)
```

```python
for word in words:
    print(word+"--->"+snowball.stem(word))
```

RegexpStemmer Class

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('$ing$|s$|e$|able$',min=4)
```

```python
print(reg_stemmer.stem("ingplaying"))
print(reg_stemmer.stem("eating"))
```
